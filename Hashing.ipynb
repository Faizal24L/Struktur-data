# Pengertian Hashing
Hashing adalah teknik dalam struktur data yang digunakan untuk menyimpan dan mencari data dengan sangat cepat. Teknik ini bekerja dengan cara mengubah sebuah nilai atau kunci (key) menjadi alamat tertentu di dalam tabel menggunakan fungsi khusus yang disebut hash function. Hasil dari proses tersebut disebut hash value atau indeks. Dengan hashing, proses pencarian data dapat dilakukan dalam waktu yang relatif singkat dibandingkan metode pencarian biasa, karena data tidak perlu dicek satu per satu.
# Fungsi Hash (Hash Function)
Fungsi hash adalah rumus atau algoritma yang digunakan untuk menentukan posisi penyimpanan data pada tabel hash (hash table). Fungsi ini menerima input berupa kunci lalu menghasilkan indeks array. Contoh sederhana fungsi hash adalah:

h(k)=kmodm
Pada rumus tersebut:
h(k) = hasil hash
k = key atau data
m = ukuran tabel
Misalnya ukuran tabel 10 dan data yang dimasukkan adalah 25, maka:
h(25)=25mod10=5

Artinya data 25 akan disimpan pada indeks ke-5
# Hash Table
Hash table adalah struktur data berbentuk tabel atau array yang digunakan untuk menyimpan data berdasarkan hasil fungsi hash. Setiap data ditempatkan pada indeks tertentu agar proses akses menjadi cepat. Struktur ini sering digunakan pada database, sistem login, kamus digital, dan cache karena efisiensinya dalam pencarian data.
# Collision pada Hashing
Collision adalah kondisi ketika dua data berbeda menghasilkan indeks hash yang sama. Hal ini sering terjadi karena ukuran tabel terbatas sedangkan jumlah data bisa banyak. Contohnya:
h(15)=15mod10=5
dan
h(25)=25mod10=5
Kedua data tersebut memiliki indeks yang sama yaitu 5 sehingga terjadi collision.
# Metode Penanganan Collision
Salah satu cara menangani collision adalah Chaining. Pada metode ini, setiap indeks tabel dapat menyimpan lebih dari satu data menggunakan linked list. Jika terjadi collision, data baru ditambahkan ke daftar pada indeks tersebut.

Metode lainnya adalah Linear Probing, yaitu mencari slot kosong berikutnya secara berurutan ketika indeks utama sudah terisi. Rumus linear probing dapat ditulis sebagai:
hi(k)=(h(k)+i)mod m
dengan i sebagai langkah pencarian slot kosong.

# Kelebihan dan Kekurangan Hashing
Kelebihan hashing adalah proses pencarian, penyisipan, dan penghapusan data menjadi sangat cepat. Struktur ini juga efisien digunakan pada data dalam jumlah besar. Namun hashing memiliki kekurangan, yaitu kemungkinan terjadinya collision serta kebutuhan ukuran tabel yang harus direncanakan dengan baik agar performa tetap optimal.


```python
def remainderFunction (data,num):
    return (data%num)
def createHashTable(num):
    temp=[]
    for i in range(num):
        temp.append('none')
    return(temp)
    
def putData(data,table):
    for i in range(len(data)):
        ind=remainderFunction(data[i],len(table))
        table[ind]=data[i]
    return(table)
    
def searchHash(data,table):
    hashVal=remainderFunction(data,len(table))
    if data==table[hashVal]:
        return True
    else:
        return False

a=[54, 26, 93, 17, 77, 31]
hashTable=createHashTable(11)
print(hashTable)

hashTable=putData(a,hashTable)
print(hashTable)

searchHash(93,hashTable)
```

```python
size = 10
tabl = [[] for _ in range(size)]

def hash_func(x):
    return x % size

def insert(table, x):
    index = hash_func(x)
    table[index].append(x)

def search(table, x):
    index = hash_func(x)
    return x in table[index]

def delete(table, x):
    index = hash_func(x)
    if x in table[index]:
        table[index].remove(x)

def update(table, hapus ,new):
    index = hash_func(hapus)
    if hapus in table[index]:
        table[index].remove(hapus)
    insert(table, new)

# UJI INSERT
data_uji = [15, 25, 35, 12, 22, 32]
for i in data_uji:
    insert(tabl, i)
    
# UJI SEARCH
print(search(tabl, 15))
print(search(tabl, 25))
print(search(tabl, 35))
print(search(tabl, 12))
print(search(tabl, 22))
print(search(tabl, 32))


# UJI DELETE
delete(tabl, 12)
delete(tabl, 35)

print(tabl)
# UJI UPDATE
update(tabl, 25, 27)
update(tabl, 22, 26)
    
print(tabl)
```
