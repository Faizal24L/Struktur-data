# struktur Data Deque
Deque, atau double-ended queue, adalah tipe struktur data yang mirip dengan queue dan stack, namun dengan kemampuan untuk menambahkan dan menghapus elemen di kedua ujungnya. Dalam deque, kita dapat menambahkan elemen baru baik di depan maupun di belakang, serta menghapus elemen dari depan dan belakang pula.

Fungsi-fungsi struktur data deque
1. Penambahan elemen
deque memnunginkan penambahan elemen baik dari depan (fornt) maupun belakang (rear)
2. penghapusan elemen
deque juga memungkinkan untuk melakukan penghapusan baik dari depan maupun belakang
3. Ukuran deque
Dengan fungsi yang mengembalikan ukuran deque kita dapat memerikasa status dequenya
4. mengekecek deque
dengan fungsi ini kita dapat mengetahu keadaan deque dalam keadaaan kosong atau  tidak
