# Linked List
Python memiliki tipe data dinamis yaitu List, yang ukurannya bisa berubah saat program berjalan dan dapat ditambah datanya kapan saja
menggunakan method bawaan. Namun, tidak semua bahasa pemrograman menyediakan tipe data dinamis seperti ini. Karena itu,
digunakan struktur data Linked List, yaitu struktur data dinamis yang dibuat sendiri oleh programmer. Hal utama dalam struktur data
Linked List:
* Node → elemen penyimpan data dan alamat node berikutnya.
* Penambahan Data → proses memasukkan node baru ke linked list.
* Traversal → proses menelusuri atau membaca seluruh isi linked list.
* Ordered List → linked list yang datanya tersusun secara terurut.
NODE Pada Linked List, data disimpan di alamat memori yang berbeda-beda. Agar data saling terhubung, setiap data membutuhkan
informasi tambahan berupa link/alamat node berikutnya. Data pertama disebut head dan harus diketahui secara eksplisit melalui external
reference. Data terakhir memiliki penanda bahwa tidak ada node berikutnya lagi.
![image.png](linkedlist.png)
Linked List tersusun dari beberapa node. Setiap node minimal memiliki:
* data → menyimpan nilai/data
* next → menyimpan alamat node berikutnya
Node biasanya dibuat menggunakan class dengan beberapa method penting:
* constructor → dijalankan saat objek dibuat
* getData() → mengambil data pada node
* getNext() → mengambil node berikutnya
* setData() → mengubah data node
* setNext() → mengatur node berikutnya

```python
# pembuatan class node
class Node:
    def __init__(self, init_data):
        self.data = init_data
        self.next = None
    def getData(self):
        return self.data
    def getNext(self):
        return self.next
    def setData(self, newdata):
        self.data = newdata
    def setNext(self, new_next):
        self.next = new_next
a=Node(93)
b=Node(20)
print(a.getNext())
print(b.getNext())
a.setNext(b)
print(a.getNext())
```

# LINKED LIST CLASS
Linked List adalah kumpulan node yang saling terhubung. Untuk mengakses seluruh node dalam linked list, harus diketahui lokasi node
pertama (head). Karena itu diperlukan pointer tambahan untuk menunjukkan keberadaan node pertama tersebut.
Class LinkedList memiliki pointer head yang menunjuk ke node pertama dalam linked list. Pada contoh juga terdapat class Node yang
memiliki data dan pointer ke node berikutnya (next). Method penting pada class LinkedList:
* constructor init → dijalankan saat objek dibuat. Karena linked list masih kosong, maka head = None.
* isEmpty() → digunakan untuk mengecek apakah linked list kosong. Jika head masih None, maka method mengembalikan nilai True.
Pada contoh kode:
* Node a dan b awalnya belum terhubung sehingga next bernilai None.
* Setelah a.setNext(b), node a akan menunjuk ke node b.
![image.png](linkedlist2.png)

```python
# kode linnked lsit class
class LinkedList:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return self.head==None
mylist=LinkedList()
print(mylist.head)
mylist.isEmpty()
print(mylist.head)
mylist.isEmpty()
```

PENAMBAHAN DATA PADA LINKED LIST

Pada Linked List, penambahan data baru biasanya dilakukan di awal list (head). Proses penambahannya terdiri dari dua langkah:

1. Node baru dihubungkan ke node pertama yang lama.
2. Pointer `head` dipindahkan agar menunjuk ke node baru.

Urutan ini tidak boleh dibalik, karena jika `head` langsung dipindah terlebih dahulu, linked list lama bisa hilang dan tidak dapat diakses lagi.

Contoh pada method `add()`:

- temp = Node(item)` → membuat node baru.
- temp.setNext(self.head)` → node baru menunjuk ke node pertama lama.
- self.head = temp` → head dipindah ke node baru.


```python
#kode penambahan method add()
class LinkedList:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return self.head==None
    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp

mylist=LinkedList()
print(mylist.head)
mylist.isEmpty()
mylist.add(34)
print(mylist.head)
mylist.add(45)
print(mylist.head)
print(mylist.head.data)
mylist.add(21)
```

TRAVERSAL LINKED LIST

Traversal adalah proses menelusuri setiap node pada Linked List menggunakan pointer bantuan bernama current, yang bergerak dari node pertama hingga node terakhir.
Traversal digunakan untuk:

- menghitung jumlah node,
- mencari data,
- menampilkan isi linked list,
- menyisipkan node,
- dan menghapus node.

Pada method size() langkah-langkahnya:

- Pointer current dimulai dari node yang ditunjuk head.
- current bergerak ke node berikutnya dengan current = current.getNext() sambil menghitung jumlah node (count).
- Proses berhenti saat current bernilai None, artinya sudah tidak ada node lagi.

```python
#kode pembuatan method size()
class LinkedList:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return self.head==None
    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp
    def size(self):
        curent = self.head
        count = 0
        while curent != None:
            count = count + 1
            curent = curent.getNext()
        return count

mylist=LinkedList()
mylist.add(45)
mylist.add(34)
mylist.add(70)
print(mylist.size())
```

PENCARIAN NODE PADA LINKED LIST

Pencarian data pada Linked List dilakukan dengan proses traversal, yaitu menelusuri node dari awal hingga akhir menggunakan pointer current. Pada setiap node dilakukan pengecekan apakah data pada node sama dengan data yang dicari, Jika data cocok, maka method search() mengembalikan nilai True. Jadi, method search() bekerja mirip dengan size(), tetapi ditambah proses pencocokan data pada setiap node.

```python
#kode methode search()
class LinkedList:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return self.head==None
    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp
    def size(self):
        curent = self.head
        count = 0
        while curent != None:
            count = count + 1
            curent = curent.getNext()
        return count
    def search(self,item):
        curent = self.head
        found = False
        while curent != None and not found:
            if curent.getData() == item:
                found = True
            else:
                curent = curent.getNext()
                return found

mylist=LinkedList()
mylist.add(45)
mylist.add(34)
mylist.add(70)
mylist.add(84)
mylist.size()
mylist.search(34)
```

PENGHAPUSAN NODE PADA LINKED LIST

Penghapusan data pada Linked List dilakukan dalam dua tahap:

- Traversal untuk mencari node yang akan dihapus.
- Remove data dengan melepas node tersebut dari linked list.

Proses traversal menggunakan pointer current untuk menelusuri node. Selain itu digunakan pointer previous yang berada satu langkah sebelum current. Saat node yang dicari ditemukan:

- previous dihubungkan ke node setelah current, sehingga node dapat dihapus tanpa merusak linked list.
- Jika node yang dihapus berada di tengah, maka previous.next menunjuk ke node setelah current.
- Jika node yang dihapus adalah node pertama (previous = None), maka pointer head langsung dipindahkan ke node berikutnya.

```python
#kode emthod remove()
class LinkedList:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return self.head==None
    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp
    def size(self):
        curent = self.head
        count = 0
        while curent != None:
            count = count + 1
            curent = curent.getNext()
        return count
    def search(self,item):
        curent = self.head
        found = False
        while curent != None and not found:
            if curent.getData() == item:
                found = True
            else:
                curent = curent.getNext()
                return found
    def display(self):
        curent = self.head
        while curent != None:
            print(curent.getData())
            curent = curent.getNext()
    def remove(self, item):
        curent = self.head
        previous = None
        found = False
        while not found:
            if curent.getData() == item:
                found = True
            else:
                previous = curent
                curent = curent.getNext()
        if previous == None:
            self.head = curent.getNext()
        else:
            previous.setNext(curent.getNext())

mylist=LinkedList()
mylist.add(45)
mylist.add(34)
mylist.add(70)
mylist.add(84)
mylist.add(97)
mylist.display()
mylist.remove(34)
mylist.display()
```

LINKED LIST PADA ORDERED LIST

Pada Linked List biasa, pencarian dilakukan dengan menelusuri node satu per satu hingga akhir. Proses ini bisa lebih cepat jika data tersusun terurut.

Ordered List adalah Linked List dengan data yang selalu dalam keadaan terurut, sehingga pencarian dapat dihentikan lebih awal saat ditemukan data yang lebih besar atau lebih kecil dari data yang dicari.

Method pada Ordered List hampir sama dengan LinkedList biasa, tetapi berbeda pada:

- add data → penambahan node harus menjaga urutan data tetap terurut.
- search data → pencarian lebih cepat karena dapat berhenti sebelum mencapai akhir list.

```python
#kode class ordered lingked list
class OrderedLinkedList:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return self.head==None
    def size(self):
        curent = self.head
        count = 0
        while curent != None:
            count = count + 1
            curent = curent.getNext()
        return count
    def display(self):
        curent = self.head
        while curent != None:
            print(curent.getData())
            curent = curent.getNext()
    def remove(self, item):
        curent = self.head
        previous = None
        found = False
        while not found:
            if curent.getData() == item:
                found = True
            else:
                previous = curent
                curent = curent.getNext()
        if previous == None:
            self.head = curent.getNext()
        else:
            previous.setNext(curent.getNext())
    def search(self,item):
        curent = self.head
        found = False
        stop=False
        while curent != None and not found and not stop:
            if curent.getData() == item:
                found = True
            else:
                if curent.getData() > item:
                    stop = True
                else:
                    curent = curent.getNext()
                    return found
    def add(self, item):
        curent=self.head
        previous = None
        stop = False
        while curent != None and not stop:
            if curent.getData() > item:
                stop = True
            else:
                previous = curent
                curent = curent.getNext()
        temp = Node(item)
        if previous == None:
            temp.setNext(self.head)
            self.head = temp
        else: # ditautkan antara previous dengan curent
            temp.setNext(curent)
            previous.setNext(temp)

myList=OrderedLinkedList()
myList.add(9)
myList.add(14)
myList.display()
myList.add(100)
myList.display()
```
