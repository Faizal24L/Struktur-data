# Linier Search
Linear Search atau pencarian linear adalah metode pencarian data yang dilakukan dengan memeriksa setiap elemen secara berurutan dari awal hingga akhir sampai data yang dicari ditemukan. Algoritma ini merupakan teknik pencarian paling sederhana dalam struktur data karena cara kerjanya hanya membandingkan satu per satu elemen pada suatu array atau list. Jika data yang dicari sama dengan elemen yang sedang diperiksa, maka pencarian berhenti dan data dinyatakan ditemukan. Namun, jika seluruh data telah diperiksa dan tidak ada yang cocok, maka data dinyatakan tidak ditemukan.
Cara kerja Linear Search dimulai dari indeks pertama pada suatu kumpulan data. Setiap elemen dibandingkan dengan nilai yang dicari menggunakan proses perbandingan sederhana. Misalnya terdapat data [10, 20, 30, 40, 50] dan pengguna ingin mencari angka 40. Program akan memeriksa angka 10 terlebih dahulu, kemudian 20, lalu 30, hingga akhirnya menemukan angka 40. Karena data ditemukan, proses pencarian langsung dihentikan. Metode ini dapat digunakan pada data yang belum terurut sehingga lebih fleksibel dibanding beberapa algoritma pencarian lainnya.
Linear Search memiliki beberapa kelebihan, yaitu mudah dipahami, mudah diimplementasikan, dan dapat digunakan pada data kecil maupun data yang belum diurutkan. Namun, algoritma ini juga memiliki kekurangan karena kurang efisien untuk jumlah data yang besar. Semakin banyak data, maka waktu pencarian akan semakin lama karena setiap elemen harus diperiksa satu per satu. Kompleksitas waktu pada Linear Search adalah O(n)O(n)O(n) pada kondisi terburuk, yaitu ketika data berada di posisi terakhir atau bahkan tidak ditemukan sama sekali. Oleh karena itu, Linear Search lebih cocok digunakan pada data dengan ukuran kecil atau saat kesederhanaan algoritma lebih diutamakan daripada kecepatan pencarian

# Binari Search
Binary Search atau pencarian biner adalah metode pencarian data yang bekerja dengan cara membagi data menjadi dua bagian secara terus-menerus hingga data yang dicari ditemukan. Algoritma ini hanya dapat digunakan pada data yang sudah terurut, baik secara ascending maupun descending. Binary Search lebih cepat dibandingkan Linear Search karena tidak memeriksa seluruh data satu per satu, melainkan langsung mempersempit area pencarian dengan membandingkan nilai tengah data.
Cara kerja Binary Search dimulai dengan menentukan elemen tengah dari suatu array atau list. Jika nilai tengah sama dengan data yang dicari, maka pencarian selesai. Jika data yang dicari lebih kecil dari nilai tengah, maka pencarian dilanjutkan pada bagian kiri array. Sebaliknya, jika data lebih besar dari nilai tengah, maka pencarian dilakukan pada bagian kanan array. Proses ini terus diulang hingga data ditemukan atau tidak ada lagi bagian data yang dapat diperiksa. Misalnya pada data [10, 20, 30, 40, 50, 60, 70] dan ingin mencari angka 50. Pertama, program memeriksa nilai tengah yaitu 40. Karena 50 lebih besar dari 40, pencarian dilanjutkan ke bagian kanan. Selanjutnya diperiksa nilai tengah baru dan ditemukan angka 50.
Binary Search memiliki kelebihan yaitu proses pencarian jauh lebih cepat dan efisien dibandingkan Linear Search, terutama pada data yang jumlahnya besar. Namun, algoritma ini memiliki kekurangan karena hanya dapat digunakan pada data yang sudah terurut. Jika data belum terurut, maka data harus diurutkan terlebih dahulu sebelum melakukan pencarian. Kompleksitas waktu Binary Search adalah O(log⁡n)O(\log n)O(logn) sehingga algoritma ini sangat efisien untuk pencarian data dalam jumlah besar. Oleh karena itu, Binary Search sering digunakan dalam berbagai sistem komputer dan aplikasi yang membutuhkan proses pencarian cepat.


```python
# sequential Search
def seqSearch(listData, data):
    ind = 0
    found = False
    while ind < len(listData) and not found:
        if listData[ind] == data:
            found = True
        else:
            ind = ind+1
    return found

a=[12,5,9,8,1,10,26]
seqSearch(a,1)

#orderedSequentialSearch
def orderedSeqSearch(listData, data):
    ind = 0
    found = False
    stop = False
    position=[]
    while ind < len(listData) and not found and not stop:
        if listData[ind] == data:
            found = True
            position.append(ind)
        else:
            if listData[ind] > data:
                stop = True
            else:
                ind = ind+1
    if found:
        return ind
    else:
        return ('Data tidak ada')
a=[1,1,5,5,5,8,9,10,12,26]
ind=orderedSeqSearch(a,5)
print(ind)
```
