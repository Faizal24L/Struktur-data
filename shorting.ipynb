# Buble Short
1. Pengertian Bubble Sort
Bubble Sort adalah algoritma pengurutan sederhana yang bekerja dengan membandingkan dua elemen bersebelahan dan menukarnya jika urutannya salah. Proses ini dilakukan berulang kali hingga tidak ada lagi pertukaran.
2. Langkah-Langkah Bubble Sort
a.Bandingkan dua elemen yang bersebelahan.
b.Tukar jika urutannya tidak sesuai.
c.Lanjutkan hingga akhir daftar.
d.Ulangi proses sampai data terurut.

Contoh Data: 5, 3, 8, 4, 2
Hasil Akhir (Ascending): 2, 3, 4, 5, 8

11. Kelebihan dan Kekurangan
Kelebihan:
a.Mudah dipahami dan diimplementasikan.
b.Cocok untuk data kecil.
c.Stabil dan tidak membutuhkan memori tambahan.

Kekurangan:
Tidak efisien untuk data besar.
Banyak melakukan pertukaran.

Kesimpulan
Bubble Sort adalah algoritma pengurutan sederhana dan mudah dipelajari.
Cocok digunakan untuk pembelajaran dasar dan dataset kecil.
Memiliki kompleksitas O(n²), sehingga kurang efisien untuk data besar.
Dapat dioptimalkan untuk meningkatkan performa pada data yang hampir terurut.

# Selection Short
1. Pengertian Selection Sort
Selection Sort adalah algoritma pengurutan sederhana yang bekerja dengan cara memilih elemen terkecil (atau terbesar) dari bagian data yang belum terurut, kemudian menukarnya dengan elemen pada posisi yang tepat.

2. Cara Kerja Selection Sort
Langkah-langkahnya:
a.Tentukan elemen pertama sebagai nilai minimum sementara.
b.Bandingkan dengan elemen lain di sebelah kanan.
c.Temukan nilai terkecil.
d.Tukar dengan posisi awal.
e.Ulangi hingga seluruh data terurut.


11. Karakteristik Selection Sort
In-place: Tidak membutuhkan memori tambahan.
Tidak Stabil: Urutan elemen yang sama bisa berubah.
Sederhana dan mudah dipahami.
Jumlah swap minimal.


12. Kelebihan dan Kekurangan

Kelebihan:
Mudah diimplementasikan.
Efisien dalam jumlah pertukaran.
Cocok untuk dataset kecil.
Tidak memerlukan memori tambahan.
Kekurangan:
Tidak efisien untuk data besar.
Selalu melakukan perbandingan sebanyak O(n²).
Tidak adaptif terhadap data yang hampir terurut.

# insertion sort

1. Pengertian Insertion Sort
Insertion Sort adalah algoritma pengurutan sederhana yang bekerja dengan cara menyisipkan elemen ke posisi yang tepat dalam bagian data yang sudah terurut.
🔹 Algoritma ini mirip dengan cara manusia menyusun kartu permainan di tangan.
🔹 Setiap elemen ditempatkan pada posisi yang benar dengan menggeser elemen lain.

2. Cara Kerja Insertion Sort
Langkah-langkahnya:
Anggap elemen pertama sudah terurut.
Ambil elemen berikutnya sebagai key.
Bandingkan dengan elemen sebelumnya.
Geser elemen yang lebih besar ke kanan.
Sisipkan key pada posisi yang tepat.
Ulangi hingga semua elemen terurut.

9. Kapan Insertion Sort Digunakan?

Insertion Sort cocok digunakan ketika:

Data berukuran kecil.
Data hampir terurut.
Membutuhkan algoritma yang stabil.
Memori terbatas.
Digunakan dalam sistem real-time atau penyisipan data secara bertahap (online).

11. Karakteristik Insertion Sort
In-place: Tidak memerlukan memori tambahan.
Stabil: Tidak mengubah urutan elemen yang sama.
Adaptif: Efisien untuk data yang hampir terurut.
Online Algorithm: Dapat mengurutkan data saat diterima.

12. Kelebihan dan Kekurangan

Kelebihan:
Mudah dipahami dan diimplementasikan.
Efisien untuk dataset kecil.
Stabil dan adaptif.
Cocok untuk pembelajaran dasar algoritma.

Kekurangan:

Tidak efisien untuk data berukuran besar.
Memerlukan banyak pergeseran elemen pada kondisi terburuk.

```python
def bubble_sort_ascending(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

data = [5, 3, 8, 4, 2]
print("Ascending:", bubble_sort_ascending(data.copy()))


def bubble_sort_descending(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] < arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

data = [5, 3, 8, 4, 2]
print("Descending:", bubble_sort_descending(data.copy()))


def bubble_sort_count_swap(arr):
    n = len(arr)
    swap_count = 0
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swap_count += 1
    return arr, swap_count


data = [5, 3, 8, 4, 2]
hasil, swap = bubble_sort_count_swap(data.copy())
print("Hasil:", hasil)
print("Jumlah Swap:", swap)
```

```python
def bubble_sort_optimized(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

data = [5, 3, 8, 4, 2]
print("Optimized:", bubble_sort_optimized(data.copy()))
```

```python
def selection_sort_ascending(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

data = [64, 25, 12, 22, 11]
print("Ascending:", selection_sort_ascending(data.copy()))

#-------------------------------------------------------------

def selection_sort_descending(arr):
    n = len(arr)
    for i in range(n - 1):
        max_index = i
        for j in range(i + 1, n):
            if arr[j] > arr[max_index]:
                max_index = j
        
        arr[i], arr[max_index] = arr[max_index], arr[i]
    return arr

data = [64, 25, 12, 22, 11]
print("Descending:", selection_sort_descending(data.copy()))

#---------------------------------------------------------------------

nama = ["Zahra", "Andi", "Budi", "Citra", "Dina"]

def selection_sort_nama(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j].lower() < arr[min_index].lower():
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

print("Urutan Alfabet:", selection_sort_nama(nama))

#------------------------------------------------------------------------

def selection_sort_optimized(arr):
    """
    Selection Sort dengan optimalisasi.
    Menemukan nilai minimum dan maksimum sekaligus
    dalam satu iterasi.
    """
    left = 0
    right = len(arr) - 1

    while left < right:
        min_index = left
        max_index = right

        # Pastikan indeks awal maksimum benar
        if arr[min_index] > arr[max_index]:
            arr[min_index], arr[max_index] = arr[max_index], arr[min_index]

        for i in range(left + 1, right):
            if arr[i] < arr[min_index]:
                min_index = i
            elif arr[i] > arr[max_index]:
                max_index = i

        # Tukar nilai minimum ke kiri
        arr[left], arr[min_index] = arr[min_index], arr[left]

        # Jika maksimum berada di posisi yang tertukar
        if max_index == left:
            max_index = min_index

        # Tukar nilai maksimum ke kanan
        arr[right], arr[max_index] = arr[max_index], arr[right]

        left += 1
        right -= 1

    return arr


# Contoh penggunaan
data = [64, 25, 12, 22, 11]
print("Selection Sort Optimized:", selection_sort_optimized(data.copy()))
```

```python
def insertion_sort_ascending(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        
        arr[j + 1] = key
    return arr

data = [8, 3, 5, 2]
print("Ascending:", insertion_sort_ascending(data.copy()))

#-------------------------------------------------------

def insertion_sort_descending(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        
        while j >= 0 and arr[j] < key:
            arr[j + 1] = arr[j]
            j -= 1
        
        arr[j + 1] = key
    return arr

data = [8, 3, 5, 2]
print("Descending:", insertion_sort_descending(data.copy()))

#---------------------------------------------------------------

nama = ["Zahra", "Andi", "Budi", "Citra", "Dina"]

def insertion_sort_nama(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        
        while j >= 0 and arr[j].lower() > key.lower():
            arr[j + 1] = arr[j]
            j -= 1
        
        arr[j + 1] = key
    return arr

print("Urutan Alfabet:", insertion_sort_nama(nama))

#-------------------------------------------------------

def insertion_sort_optimized(arr):
    """
    Insertion Sort dengan optimalisasi.
    Menggunakan pergeseran elemen dan berhenti lebih awal
    jika posisi elemen sudah benar.
    """
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        # Jika elemen sudah pada posisi yang tepat, lewati
        if arr[j] <= key:
            continue

        # Geser elemen yang lebih besar ke kanan
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        # Sisipkan elemen pada posisi yang benar
        arr[j + 1] = key

    return arr


# Contoh penggunaan
data = [64, 25, 12, 22, 11]
print("Insertion Sort Optimized:", insertion_sort_optimized(data.copy()))
```
